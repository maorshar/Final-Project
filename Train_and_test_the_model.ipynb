```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import shutil
import os
```

The model and the training

```python
# Define the paths to the color block and traffic sign image directories
color_block_dir = "/content/extracted_folder/content/traffic_sign_AE/"
traffic_sign_dir = "/content/extracted_folder2/content/traffic_sign_AE_seg/"
```

```python
from keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Input, Concatenate
from keras.models import Model
import keras.backend as K
from keras.callbacks import EarlyStopping

K.clear_session()

img_height = 128
img_width =  128
batch_size = 128

def paired_generator(color_block_generator, traffic_sign_generator):
    while True:
        color_block_images = color_block_generator.next()[0]
        traffic_sign_images = traffic_sign_generator.next()[0]
        yield (traffic_sign_images, color_block_images)  #swapped

# Use the Keras ImageDataGenerator to load the images and create a paired dataset
data_generator = keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
color_block_dataset = data_generator.flow_from_directory(
    directory=color_block_dir,
    target_size=(img_height, img_width),
    color_mode="rgb",
    batch_size=batch_size,
    class_mode="input",
    shuffle=True,
    seed=42,

)
traffic_sign_dataset = data_generator.flow_from_directory(
    directory=traffic_sign_dir,
    target_size=(img_height, img_width),
    color_mode="rgb",
    batch_size=batch_size,
    class_mode="input",
    shuffle=True,
    seed=42
)

# Merge the paired datasets using the generator function
paired_dataset = paired_generator(color_block_dataset,traffic_sign_dataset)
# print(list(paired_dataset))

# Define the autoencoder model
input_shape = (img_height, img_width, 3)
inputs = Input(shape=input_shape)
x = Conv2D(128, (5, 5), activation='relu', padding='same')(inputs)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (5, 5), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(32, (5, 5), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(16, (5, 5), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(16, (5, 5), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (5, 5), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(64, (5, 5), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(128, (5, 5), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = Model(inputs, decoded)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')

# Train the model
total_samples = color_block_dataset.n
steps_per_epoch = total_samples // batch_size
# Define the early stopping callback
early_stopping = EarlyStopping(monitor='loss', patience=8, restore_best_weights=True)

history = autoencoder.fit(paired_dataset, epochs=140, steps_per_epoch=steps_per_epoch, callbacks=[early_stopping])

import matplotlib.pyplot as plt

# Plot the loss graph
plt.plot(history.history['loss'])
plt.title('Autoencoder Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
